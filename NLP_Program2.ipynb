```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df1=pd.read_csv('/content/SMS_test.csv',encoding='ISO-8859-1')
df1
```

```python
df2=pd.read_csv('/content/SMS_train.csv',encoding='ISO-8859-1')
df2
```

```python
df=pd.concat([df2,df1],ignore_index=True)
df
```

```python
df.drop(['S. No.'],axis=1,inplace=True)
df
```

```python
import seaborn as sns
sns.countplot(x='Label',data=df)
```

```python
df['Label']=df['Label'].map({'Non-Spam':1,'Spam':0})
df
```

```python
msg=df.Message_body
msg
```

```python
msg=msg.str.replace('[^a-zA-Z0-9]+'," ")
msg
```

```python
import nltk
from nltk.stem import SnowballStemmer
from nltk import word_tokenize
nltk.download('punkt')
stemmer=SnowballStemmer('english')
```

```python
msg=msg.apply(lambda x:[stemmer.stem(i.lower())for i in word_tokenize(x)]).apply(lambda y:" ".join(y))
msg
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
sw=stopwords.words('english')
sw
```

```python
msg=msg.apply(lambda x:[i for i in word_tokenize(x) if i not in sw]).apply(lambda y:" ".join(y))
msg
```

```python
msg=msg.apply(lambda x:[i for i in word_tokenize(x) if len(i)>=3]).apply(lambda y:" ".join(y))
msg
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
train_data=vectorizer.fit_transform(msg)
```

```python
X=train_data
y=df['Label'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
sv=SVC()
mn=MultinomialNB()
rf=RandomForestClassifier()
ab=AdaBoostClassifier()
lst=[sv,mn,rf,ab]
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay
for i in lst:
  print(i)
  i.fit(X_train,y_train)
  y_pred=i.predict(X_test)
  print(classification_report(y_test,y_pred))
  print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
